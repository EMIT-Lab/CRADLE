```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# Load iris dataset as an example
df_aug=pd.read_csv('./dataset_augmented_all.csv')
df_ini=pd.read_csv('../processed_data.csv')
```

```python
input=df_ini.drop(columns =['size_nm','g_factor*10^4'], inplace = False, axis = 1)
output=df_aug[['size_nm','g_factor*10^4']]
```

```python
fig, (ax1,ax2,ax3,ax4) =plt.subplots(1,4, figsize=(18,5))
fontsize=15
ax1.plot(input['num of group'],output['g_factor*10^4'],'ob')
ax1.set_xlabel('number of groups',fontsize=fontsize)
ax1.set_ylabel('g_factor*10^4',fontsize= fontsize)

ax2.plot(input['spcaer'],output['g_factor*10^4'],'oc')
ax2.set_xlabel('spacers',fontsize=fontsize)
ax2.set_ylabel('g_factor*10^4',fontsize= fontsize)

ax3.plot(output['size_nm'],output['g_factor*10^4'],'om')
ax3.set_xlabel('size (nm)',fontsize=fontsize)
ax3.set_ylabel('g_factor*10^4',fontsize= fontsize)

ax4.plot(input['time_min'],output['g_factor*10^4'],'or')
ax4.set_xlabel('time (min)',fontsize=fontsize)
ax4.set_ylabel('g_factor*10^4',fontsize= fontsize)


fig.tight_layout()
fig.show()
fig.savefig('plot.svg',dpi=150,format='svg')
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,10),dpi=150)
cax=ax.scatter( input['metal_amount_mmol'],input['ligand_amount_mmol'] , marker='o', c=output['g_factor*10^4'], cmap='winter')
plt.colorbar(cax, label='g_factor*10^4')
#ax.set_title('',size=15)
plt.xlim((-0.1,1.4))
plt.ylim((-0.1,5))
plt.xlabel('metal_amount_mmol',fontsize=20)
plt.ylabel('ligand_amount_mmol',fontsize=20)
fig.show()
#fig.savefig('PCA.svg',dpi=150,format='svg')
```

```python
fig, ax =plt.subplots(1,1, figsize=(20,5))
ligand_metal_ratio = input['ligand_amount_mmol']/input['metal_amount_mmol']
ax.plot(ligand_metal_ratio, output['g_factor*10^4'],'o')
plt.xlim(-0.1,5)
plt.show()
```
